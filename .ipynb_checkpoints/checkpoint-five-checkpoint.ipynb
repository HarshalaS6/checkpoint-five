## Prepare Your Dataframe

Import any libraries you need and create a dataframe.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
import statsmodels.api as sm

cleaned_df = pd.read_csv("cleaned_weather_repository.csv")

cleaned_df.info()
```

## Find Correlations

Use either pairplot or a heatmap or both to find the two variables with the strongest correlation in your dataset.

```python
# Using heat map visualize the data for correlation
sns.pairplot(cleaned_df, x_vars=['humidity', 'temperature_fahrenheit'],
                y_vars='condition_text', height=4,aspect=1, kind='scatter')
plt.show()
```

```python
#heatmap for correlation
sns.heatmap(cleaned_df.corr(numeric_only = True), cmap="YlGnBu", annot = True)
plt.show()
```

```python
#creating visualization using heat map for correlation between uv_index and feels_like_fahrenheit


plt.title("correlation between UV Index and Temp Fahrenheit")
sns.heatmap(cleaned_df[["humidity","feels_like_fahrenheit"]].corr(), annot = True)
plt.show()
```

## Create Your Model
Use the two columns with the strongest correlation to create and train your model. Make sure to print out the summary and plot the column values and the line produced by the model.

```python
X = cleaned_df['humidity']
y = cleaned_df['feels_like_fahrenheit']


X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, 
                                                    test_size = 0.3, random_state = 100)

X_train_sm = sm.add_constant(X_train)

lr = sm.OLS(y_train, X_train_sm).fit()

lr.params
lr.summary()
```

```python
#[2] The condition number is small 290 which indicate that there is a multicollinearity.
```

```python
plt.scatter(X_train, y_train)
plt.plot(X_train, 84.0494 + -0.01211*X_train, 'r')
plt.show()
```
